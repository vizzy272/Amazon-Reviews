```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("AmazonReviews") \
    .config("spark.sql.caseSensitive", "true") \
    .getOrCreate()

file_path = "/content/drive/MyDrive/Electronics.json.gz"

df_raw = spark.read.json(file_path)

print("Total rows:", df_raw.count())
print("Columns:", df_raw.columns)
```

```python
from pyspark.sql.functions import col

df_raw = df_raw.withColumn("label", col("overall").cast("int"))

df_clean = df_raw.filter(col("reviewText").isNotNull())

print("Clean rows:", df_clean.count())
```

```python
df_small = df_clean.limit(50000)

print("Limited rows:", df_small.count())
```
